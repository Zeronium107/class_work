```python
import numpy as np
import matplotlib.pyplot as plt

#Load wavefunction data
data = np.loadtxt("wavefunction_0.txt")
x = data[:, 0] 
psi = data[:, 1]

#constants
V0 = 50 * 1.6022e-19    #50 eV in joules
a = 1e-11  #a = 10^-11 m

#Compute potential V(x) in eV
V = V0 * (x**2) / (a**2) / 1.6022e-19  # back to eV

#Scale psi for visibility since it could not be seen
psi_scaled = psi * 100

#Plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x * 1e10, psi_scaled, label=r"$\psi(x)$ (scaled)")  # x in 
ax.plot(x * 1e10, V, label="V(x) [eV]", color='orange')

ax.set_xlabel("x []")
ax.set_ylabel("Amplitude / Energy [eV]")
ax.set_title("Wavefunction and Harmonic Potential")
ax.legend()
ax.grid(True)


plt.show()
```

```python
# since the potential and the wavefunction dont seem to scale properly we just create two axis
fig, ax1 = plt.subplots(figsize=(10, 6))

for i in range(3):
    data = np.loadtxt(f"lab/wavefunction_{i}.txt")
    x = data[:, 0]
    psi = data[:, 1]
    ax1.plot(x * 1e10, psi, label=r"$\psi(x)$")

# First y-axis for psi(x)
ax1.set_xlabel("x")
ax1.set_ylabel(r"$\psi$")
ax1.tick_params(axis='y')

# Second y-axis for potential
ax2 = ax1.twinx()
ax2.set_ylabel("V(x)", color="orange")
ax2.plot(x * 1e10, V, color="orange", label="V(x)")
ax2.tick_params(axis='y', labelcolor="orange")

fig.suptitle("Wavefunction and Potential")
fig.tight_layout()
plt.grid(True)
plt.show()
```
