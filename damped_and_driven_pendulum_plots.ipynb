# Lab a theta = 0.314

```python
fig, axs = plt.subplots(1, 2, figsize = (15,10))
data = np.loadtxt('pendulum.txt', skiprows=1)
t = data[:, 0]
theta = data[:, 1]
w = data[:, 2]
axs[0].scatter(t, theta, s = 1)
axs[0].set_xlabel("t")
axs[0].set_ylabel(r"$\theta$")
axs[0].set_title(r"$\theta$" " Vs t", fontsize = 10)

axs[1].scatter(theta, w, s = 1)
axs[1].set_xlabel(r"$\theta$")
axs[1].set_ylabel("w")
axs[1].set_title("w vs " r"$\theta$", fontsize = 10)

fig.suptitle("Lab 1a Pendulum: " + r"$\theta$" " = 0.314", fontsize=14)
#plt.savefig("/home/jnzeron/class_notes/homework/HW_13/figs/pendulum_lab1a.pdf", bbox_inches='tight')
plt.show()
```
