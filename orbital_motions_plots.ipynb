# Earth Jupiter

```python
fig, axs = plt.subplots(figsize = (10,6))
data = np.loadtxt('orbit_earth_jupiter.txt', skiprows=2)
t = data[:, 0]
re = data[:, 5]
ye = data[:, 1]
xe = data[:, 2]

rp = data[:, 10]
yp = data[:, 6]
xp = data[:, 7]

axs.scatter(xe, ye, label="Earth", s = 10)
axs.scatter(xp, yp, label="Jupiter", s = 10)

axs.set_xlabel("x")
axs.set_ylabel("y")
plt.title("Earth Jupiter Orbit")

plt.legend(loc = 'upper right')
plt.show()
plt.savefig("/home/jnzeron/class_notes/homework/HW_11/figs/HWa_orbits.pdf", bbox_inches='tight')
```
